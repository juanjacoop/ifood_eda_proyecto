Abajo, busco Outliers con distintos boxplots, no es necesario correr la celda de abajo.

```python
for column in df.columns:
    if df[column].dtype == np.int64 or df[column].dtype == np.float64:  
        df[column].plot.box()
        plt.title(f'{column}')
        plt.show()
```

```python
def interquartile_outlier_drop(data_frame):
    Q1 = data_frame.quantile(0.25)
    Q3 = data_frame.quantile(0.75)
    IQR = Q3 - Q1

    data_frame = data_frame[~((data_frame < (Q1 - 1.5 * IQR)) |(data_frame > (Q3 + 1.5 * IQR)))]

    data_frame = data_frame.dropna()
    return data_frame
```

# EJERCICIO 1

¿Existe una relación entre los ingresos anuales de los clientes(`income`) y la cantidad
de dinero(Suma de las columnas `Mnt..`) que gastan en iFood? 

Visualiza la relación con un gráfico y
describe tus conclusiones

```python
df['MntTotal'] = df[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)
df_ejerc_1 = df[['Income','MntTotal']]

df_ejerc_1_limpio = interquartile_outlier_drop(df_ejerc_1)

df_ejerc_1_limpio.info()
```

```python
x = df_ejerc_1['MntTotal']
y = df_ejerc_1['Income']
plt.scatter(x, y)
plt.title ('Datos sin limpiar')
plt.xlabel('Gasto en Ifood')
plt.ylabel('Ingreso')
plt.show() 

x = df_ejerc_1_limpio['MntTotal']
y = df_ejerc_1_limpio['Income']
plt.scatter(x, y)
plt.title ('Datos limpios')
plt.xlabel('Gasto en Ifood')
plt.ylabel('Ingreso')
plt.show() 
```

A pesar de que hay un outlier de un ingreso muy alto, podemos ver que la tendencia de los puntos indica una relación entre gasto e ingreso podemos confirmar esto con una matriz de correlacion

```python
print("Correlacion de datos sin limpiar")
df_ejerc_1.corr()
```

```python

print("Correlacion de datos limpios")
df_ejerc_1_limpio.corr()
```

Podemos ver una correlacion positiva entre gasto e ingreso.

# EJERCICIO 3

¿Cómo varía el tipo de gasto en comida según los ingresos de los clientes?

Muestra visualmente que tipo de productos son los más consumidos por los clientes según su nivel de ingresos

```python
df["Income"].hist()
plt.show()

df_ej_3 = df[['Income','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']]
df_ej_3_limpio = interquartile_outlier_drop(df_ej_3)

df_ej_3_limpio["Income"].hist()
plt.show()

df_ej_3_limpio.info()
```

Vemos ahora, como quitando outliers, nuestra distribucion de ingresos se ajusta a una distribucion normal

```python
df_ej_3_limpio.describe()

df_ej_3_limpio["Nivel_Ingresos"] = pd.qcut(df_ej_3_limpio['Income'], q=3, labels=['Bajo','Medio','Alto']) #Separamos los ingresos por quartil. q=3 indica que creamos 3 cuartiles distintos


df_ej_3_limpio.head()



```

```python
df_ej_3_final = df_ej_3_limpio[['Nivel_Ingresos','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']]

df_ej_3_final.head()
```

```python
df_ej_3_final.groupby('Nivel_Ingresos').sum().plot(kind='bar')
```
